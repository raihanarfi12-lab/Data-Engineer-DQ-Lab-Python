## Pengenalan tipe data

Selain itu terdapat beberapa type data seperti array,tupple,set,dan dictionary. Pada code di bawah,kita bisa lihat tipe data pada suatu variabel dengan type(var)

```python
#Untuk melihat tipe data
var_string="Belajar Python DQLAB"
var_int=10
var_float=3.14
var_list = [1,2,3,4]
var_tuple=("satu","dua","tiga")
var_dict = {"nama":"Raihan","umur":20}

print(var_string)
print(var_int)
print(var_float)
print(var_list)
print(var_tuple)
print(var_dict)

print(type(var_string))
print(type(var_int))
print(type(var_float))
print(type(var_list))
print(type(var_tuple))
print(type(var_dict))
```

## Percabangan

Percabangan biasanya digunakan untuk melakukan instruksi berdasarkan kondisi tertentu. Kita dapat menuliskan if(kondisi1) untuk menerapkan instruksi jika kondisi1 terjadi. Jika tidak terjadi, maka kita dapat menggunakan else

```python
i=2
if(i<7):
	print("nilai i kurang dari 7")
  # Dalam if,dapat pula diletakkan if lagi di dalamnya. Hal ini biasanya disebut nested if
	if(i<3):
		print("nilai i kurang dari 7 dan kurang dari 3")
	else:
		print("nilai i kurang dari 7 tapi lebih dari 3")
```

Jika kondisi nya cukup banyak, kita dapat menggunakan elif

```python
i =10 #dapat diganti dengan angka lain

if(i==5):
	print("ini adalah angka 5")
elif(i>5):
	print("lebih besar dari 5")
else:
	print("lebih kecil dari 5")
```

## Perulangan

Perulangan dilakukan untuk mengulang kembali suatu instruksi hingga memenuhi kondisi tertentu (while) atau pada rentang tertentu(for)

```python
#contoh penerapan while
j=0
while j<6:
	print("Ini adalah perulangan ke -",j)
	j=j+1
```

```python
for i in range (1,11):
    if(i%2==0):
        print("Angka genap",i)
    else:
         print("Angka ganjil",i)
```

## Fungsi
Fungsi merupakan pada dasarnya suatu kumpulan perintah atau instruksi. Kegunaan fungsi adalah agar kita tidak perlu menulis code dengan tujuan yang sama secara berulang. Misalnya kita ingin menghitung luas, maka kita dapat menggunakan fungsi luas yang sudah ditulis pada code 

```python
#pada fungsi di bawah, alas dan tinggi merupakan parameter yang digunakan fungsi. Selain itu return berfungsi nilai yang ingin kita kembalikan
def luas_segitiga(alas,tinggi):
	luas=(alas*tinggi)/2
	return luas

print("Luas segitiga: %d" % luas_segitiga(4,6))

def luas_segitiga(alas,tinggi):
	luas=(alas*tinggi)/2
	return alas,tinggi,luas

print("Alas,tinggi,dan luas segitiga: %d,%d,%d" % (luas_segitiga(4,6)))
```

## Matplotlib

Matplotlib adalah package yang digunakan untuk visualisasi data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


table = pd.read_csv("https://storage.googleapis.com/dqlab-dataset/penduduk_gender_head.csv")
table.head()

x_label = table['NAMA KELURAHAN']
plt.bar(x=np.arange(len(x_label)),height=table['LAKI-LAKI WNI'])
plt.xticks(np.arange(len(x_label)),table['NAMA KELURAHAN'],rotation=90)
plt.xlabel('Kelurahan di Jakarta Pusat')
plt.ylabel('Jumlah Penduduk Laki - Laki')
plt.title('Persebaran Jumlah Penduduk Laki- Laki di Jakarta Pusat')

plt.show()
```
